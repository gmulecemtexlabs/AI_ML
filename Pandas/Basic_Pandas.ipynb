## Series
A Pandas Series is like a column in a table.
It is a one-dimensional array holding data of any type.

```python
myList = [2, 4, 5, 56]

y = pd.Series(myList)
print(y)
```

## Lables
With the index argument, you can name your own labels.

```python
myList = [2, 4, 5, 56]

y = pd.Series(myList, index = ['a', 'b', 'c', 'd'])
print(y)
```

```python
y['d']
```

```python
y['b']
```

## DataFrames
Data sets in Pandas are usually multi-dimensional tables, called DataFrames.
Series is like a column, a DataFrame is the whole table.

```python
mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2]
}

df = pd.DataFrame(mydataset)
print(df)
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data)
df
```

**Pandas use the loc attribute to return one or more specified row(s)**

```python
df.loc[[1, 2]]
```

# Cleaning data

```python
df = pd.read_csv("sample.csv")

new_df = df.dropna()
new_df.head()
```

### 1. Remove empty rows

```python
df = pd.read_csv("sample.csv")

df.dropna()
df.head()
```

#### If you want to change the original DataFrame, use the **inplace = True** argument:

```python
df = pd.read_csv("sample.csv")
df.dropna(inplace = True)
df.head()
```

### 2. Replace Empty Values
Another way of dealing with empty cells is to insert a new value instead.

```python
df = pd.read_json("JSON_data.json")
df.fillna("India", inplace = True)
df.head()
```

### 3. Replace Only For Specified Columns
The example above replaces all empty cells in the whole Data Frame.

```python
df =  pd.read_json("JSON_data.json")
df["country"].fillna("India", inplace = True)
df.head()
```

```python
df =  pd.read_json("JSON_data.json")
df['creditBalance'].fillna(100, inplace = True)
df.head()
```
